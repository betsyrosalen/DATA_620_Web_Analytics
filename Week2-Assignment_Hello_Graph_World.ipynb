# Week 2 - Assignment Hello Graph World

First import necessary packages for plotting graphs using NetworkX and Matplotlib  and set up graph size parameters...

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10,8)
```

Create a graph object...

```python
g = nx.Graph()
```

Add nodes (or vertices) to graph object...

```python
g.add_nodes_from(["Andre","Beverly","Ed","Carol","Diane","Fernando","Garth","Heather","Ike","Jane"])
```

Create edges (representing relationships) between nodes...

```python
g.add_edges_from([("Andre","Beverly"),("Andre","Carol"),("Andre","Diane"),("Andre","Fernando"),
                  ("Beverly","Diane"),("Beverly","Ed"),("Beverly","Garth"),
                  ("Carol","Diane"),("Carol","Fernando"),
                  ("Diane","Ed"),("Diane","Fernando"),("Diane","Garth"),
                  ("Ed","Garth"),
                  ("Fernando","Garth"),("Fernando","Heather"),
                  ("Garth","Heather"),
                  ("Heather","Ike"),
                  ("Ike","Jane")])
```

Draw the graph...

```python
nx.draw(g, with_labels=True)
```

Now let's play around with the different ways of drawing the graph to seee what the differences are:

First let's see what happens when we use the `draw_networkx` method instead of just the `draw` method...

```python
nx.draw_networkx(g)
```

Not a whole lot of difference there.  Biggest difference is that it automatically puts a solid line border around the image.  

Let's play around with some of the `draw_networkx` method options...

```python
nx.draw_networkx(g, node_size=1000, node_color="y", font_size=12, font_family="calibri", 
                 width=2, edge_color="darkgray")
```

That's a bit better, it's got nicer colors and it's easier to read. I found out from the documentation that it defaults to the `draw_spring` layout, which is better than the default layout of the `draw` method.  I could not figure out how to get rid of that border though!

Let's see what the other layout methods look like...

### `draw_circular` layout

```python
nx.draw_circular(g, with_labels=True, node_size=1000, node_color="y", font_size=12, font_family="calibri", 
                 width=2, edge_color="darkgray")
```

Not bad but not the optimal layout for this particular graph.

### `draw_random` layout

```python
nx.draw_random(g, with_labels=True, node_size=1000, node_color="y", font_size=12, font_family="calibri", 
                 width=2, edge_color="darkgray")
```

Definitely not an optimal layout, and you never know what you are going to get if you run it again!

### `draw_spectral` layout

```python
nx.draw_spectral(g, with_labels=True, node_size=1000, node_color="y", font_size=12, font_family="calibri", 
                 width=2, edge_color="darkgray")
```

Not sure what kind of graph you would want to use this layout for?  Why would you want nodes to be drawn on top of each other?

### `draw_spring` layout

```python
nx.draw_spring(g, with_labels=True, node_size=1000, node_color="y", font_size=12, font_family="calibri", 
                 width=2, edge_color="darkgray")
```

This is the same as the `draw_networkx` method but no border!

### `draw_shell` layout

```python
nx.draw_shell(g, with_labels=True, node_size=1000, node_color="y", font_size=12, font_family="calibri", 
                 width=2, edge_color="darkgray")
```

This looks exactly like the `draw_circular` method.  Not sure what the difference is?
