### Week 3 - Graph Visualization

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (15,15)
```

```python
dolphins = nx.read_edgelist('mammalia-dolphin-florida-social/mammalia-dolphin-florida-social.edges', nodetype=int, 
                                     data=(('weight',int),))
# Removed opne line from the edges file "37 85 1" beccause there were no other connections between those two dolphins 
# and the rest of the graph and it made the chart hard to see all at once.
```

```python
nx.draw(dolphins, with_labels=True, node_color="y", node_size=400, font_weight='bold', edge_color="darkgray")
```

```python
macaques = nx.read_edgelist('mammalia-macaque-dominance/mammalia-macaque-dominance.edges', nodetype=int, 
                                     data=(('weight',int),))
```

```python
nx.draw(macaques, with_labels=True, node_color="y", node_size=1000, font_weight='bold', edge_color="darkgray")
```
